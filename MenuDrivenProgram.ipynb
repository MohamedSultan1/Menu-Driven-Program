```python
print("WELCOME TO MY MENU DRIVEN PROGRAM\n") # Display a Welcome msg

# creating list from user input

print("Please Enter a List of Numbers Below: ") # Asking user to enter a list
list = [int(x) for x in input().split()] # Creating a global variable called list to store the list entered by user
print(list)

def Insert(list_temp): # Initializing Insert Function with a local variable called list_temp
        print("\nINSERTING\n") # Displaying Isert msg
        print("") # making a space

        insertNum = int(input("Enter a number to be inserted: ")) # Asking user to enter a number to be inserted
        list_temp.append(insertNum) # Appending (adding) the entered number to our list
        print("The list after number inserted: ", list_temp) # Displaying the list after the number has been inserted
        return list_temp


def Remove(list_temp): # Initializing Remove Function with a local variable called list_temp
        print("\nREMOVING\n") # Displaying Remove msg
        print("") 

        removeNum = int(input("Enter a number to be removed: ")) # Asking user to enter a number to be removed
        list_temp.remove(removeNum) # removing the entered number from our list
        print("The list after number removed: ", list_temp) # Displaying the list after the number has been removed
        return list_temp

def Sort(list_temp): # Initializing Sort Function with a local variable called list_temp
        print("\nSORTING\n") # Displaying Sort msg
        print("") 

        list_temp.sort() # sorting the numbers in our list
        print("The list after sorting: ", list_temp) # Displaying the list after it has been sorted
        return list_temp    

def Extend(list_temp): # Initializing Extend Function with a local variable called list_temp
        print("\nEXTENDING\n") # Displaying Extend msg
        print("") 
        extendNum = [int(x) for x in input("Enter a list of numbers to be extended: ").split()] # Asking user to enter a number to be extended
        list_temp.extend(extendNum) # Extending the entered number to our list
        print("The list after extending: ", list_temp) # Displaying the list after the number has been extended
        return list_temp

def Reverse(list_temp): # Initializing Reverse Function with a local variable called list_temp
        print("\nREVERSING\n") # Displaying Reverse msg
        print("") 

        list_temp.reverse() # reversing the numbers in our list
        print("The list after reversing: ", list_temp) # Displaying the list after it has been reversed
        return list_temp 

def Traverse(list_temp): # Initializing Traverse Function with a local variable called list_temp
        print("\nTRAVERSING\n") # Displaying Traverse msg
        print("") 

        searchNum = int(input("Enter a number to be searched for: ")) # Asking user to enter a number to be searched for
        for x in (list_temp): # for loop iterating through our list to find inputed number
          if x == searchNum:
            print("Number FOUND!", searchNum) # if the for loop finds the number, display Found!
            return

        print("Number NOT FOUND!", searchNum) # if the for loop does NOT find the number, Display NOT Found!
        return



def Operations():
# creating options function called Operations

# While loop displaying all our options (7 Options)
  while True:
    print("\nMAIN MENU\n")
    print("1. Insert")
    print("2. Remove")
    print("3. Sort")
    print("4. Extend")
    print("5. Reverse")
    print("6. Traverse")
    print("7. Exit")
    user_choice = int(input("Enter your choice: ")) # asking user to enter their choice


    if user_choice == 1: # if user chooses option 1, call Insert function
        Insert(list)
        
    elif user_choice ==2: # if user chooses option 3, call Remove function
        Remove(list)

    elif user_choice ==3: # if user chooses option 3, call Sort function
        Sort(list)

    elif user_choice ==4: # if user chooses option 4, call Extend function
        Extend(list)

    elif user_choice ==5: # if user chooses option 5, call Reverse function
        Reverse(list)

    elif user_choice ==6: # if user chooses option 6, call Traverse function
        Traverse(list)


    elif user_choice ==7: # if user chooses option 7, print the final list and exit
        print(list)
        return

        Exit = input("Press N to Exit. ") # to Exit from a specific option, press n or N
        if Exit in 'Nn':
          break  

    else:
        print("Oops! Incorrect Choice.") # if non of previous choices is used, display incorreect choice
        break
Operations() # calling our Operations function

    
```
